```python
# Création d'une variable 'loyer_moyen' au m2.

df['loyer_moyen'] = round((df['max'] + df['min']) /2, 2)
```

```python
# Choix des variables pour le ML.

df_ml = df[['annee', 'nom_quartier', 'piece', 'meuble_txt', 'epoque', 'loyer_moyen']]
```

```python
df_ml
```

```python
df['annee'] = df['annee'].astype(str)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Initialisation du X et du y (variable cible).
X = df_ml.drop(['loyer_moyen'], axis=1)
y = df_ml['loyer_moyen']

# Get_dummies sur le X.
X = pd.get_dummies(X).astype(int)

# Train test split.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

# Entrainement du model de linearRegression.
model = LinearRegression().fit(X, y)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
future_years = list(range(2024, 2055))   # Génère les année à prédire de 2024 à 2054 inclus.
quartiers = df_ml['nom_quartier'].unique() # Valeur unique des noms de chaque quartier.
pieces = df_ml['piece'].unique() # Valeur unique du nombre de pièce.

# Création d'une liste vide pour stocker les données de predictions.
future_data_list = []


# Combinaison de quartier, piece, et year, où un dictionnaire est créé et ajouté à la liste 'future_data_list'.
for quartier in quartiers:
    for piece in pieces:
        for year in future_years:
            future_data_list.append({
                'nom_quartier': quartier,
                'piece': piece,
                'annee': year
            })

# Liste en DF
future_data_df = pd.DataFrame(future_data_list)

# Get dummies sur 'future_data_df et reindex pour avoir le même nombre de colonne que X, pour faire un predict.
future_data = pd.get_dummies(future_data_df).astype(int)
future_data = future_data.reindex(columns=X.columns, fill_value=0)

# Prédire les valeurs pour les 10 prochaines années
predictions = model.predict(future_data)

# Combiner les résultats dans un DataFrame
future_df = pd.DataFrame({
    "nom_quartier": future_data_df['nom_quartier'],
    "piece": future_data_df['piece'],
    "year": future_data_df['annee'],
    "predicted_value": predictions
})

future_df

```

```python
# Pivot table afin d'avoir les années en colonne.

future_df_annee = pd.pivot_table(future_df, index = ['nom_quartier', 'piece'], values='predicted_value', columns=['year']).reset_index()
future_df_annee
```

```python
# Calcul des taux de variation tous les 10 ans.


future_df_annee['variation_%__2034'] = round(((future_df_annee[2034] - future_df_annee[2024]) / future_df_annee[2024])*100,2)
future_df_annee['variation_%__2044'] = round(((future_df_annee[2044] - future_df_annee[2024]) / future_df_annee[2024])*100,2)
future_df_annee['variation_%__2054'] = round(((future_df_annee[2054] - future_df_annee[2024]) / future_df_annee[2024])*100,2)
```
